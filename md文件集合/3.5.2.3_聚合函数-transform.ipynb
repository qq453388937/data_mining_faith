```python
from pandas import DataFrame, Series
import pandas as pd
import numpy as np

data_frame = DataFrame(
    np.arange(36).reshape((6, 6)), 
    columns=list('abcdef')
)

data_frame['key'] = Series(list('aaabbb'), name='key')

data_frame
```

```python
# 对数据进行分组
group_data_frame = data_frame.groupby('key')

# 打印分组数据
dict([x for x in group_data_frame])['a']
```

```python
dict([x for x in group_data_frame])['b']
```

```python
# transform实现
# transform对组内的各个元素执行组级别的操作. 最终返回一个聚合结果.
def my_func(x):
    return x - x.mean()

group_data_frame.transform(my_func)
group_data_frame.transform(np.mean).add_prefix('mean_')
```
