```python
from pandas import DataFrame,Series
import pandas as pd
import numpy as np

data_frame = DataFrame(
    np.random.randint(100, 1000, (10, 10)), 
    index=[ 'key' + str(ix) for ix in range(1, 11)], 
    columns=[ 'data' + str(ix) for ix in range(1, 11)]
)

data_frame
```

```python
# 将表格代码写入文件
with open('data_frame.html', 'w') as f:
    f.write(data_frame.to_html())
```

```python
# 抓取本地HTML文件中的表格
pd.read_html('data_frame.html')[0]
```
