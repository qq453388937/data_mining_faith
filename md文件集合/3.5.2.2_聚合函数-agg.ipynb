```python
from pandas import DataFrame, Series
import pandas as pd
import numpy as np

data_frame = DataFrame(
    np.arange(36).reshape((6, 6)), 
    columns=list('abcdef')
)

data_frame['key'] = Series(list('aaabbb'), name='key')
data_frame
```

```python
# 对数据分组
data_group = data_frame.groupby('key')
# 输出分组数据信息
dict([x for x in data_group])['a']
```

```python
dict([x for x in data_group])['b']
```

```python
# agg和应用单个函数
data_group.agg('mean')
data_group.mean()
```

```python
# 使用agg对每一列数据分别应用多个函数
data_group.agg(['mean', 'sum'])
```

```python
# 我们可以自定义列名
data_group.agg([('平均值', 'mean'), ('总和', 'sum')])
```

```python
# 我们也可以指定不同的列应用不同的函数
# 例如data1使用sum、mean, data2使用mean
data_group.agg({'a': [('总和', 'sum'), ('平均值', 'mean')], 'b': 'mean'})
```
