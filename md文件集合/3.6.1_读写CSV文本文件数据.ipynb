```python
from pandas import DataFrame,Series
import pandas as pd
import numpy as np

# 创建pandas对象

series = Series(
    np.random.randint(100, 1000, 10), 
    index=['key' + str(ix) for ix in range(1, 11)]
)

data_frame = DataFrame(
    np.random.randint(100, 1000, (10, 10)), 
    index=[ 'key' + str(ix) for ix in range(1, 11)], 
    columns=[ 'data' + str(ix) for ix in range(1, 11)]
)

# 设置行列名字
data_frame.index.name = 'key'

data_frame
```

```python
series
```

```python
# 将其写入csv文件, 参数为文件名
series.to_csv('series.csv')
data_frame.to_csv('data_frame.csv')
```

```python
# 从csv文件中读取数据
# header=0表示第一行作为列索引 header=None表示没有列索引
# index_col=0表示使用第一列作为行索引
pd.read_csv('series.csv', index_col=0, header=None)
```

```python
# 默认会把第一行作为列索引
# 行索引如果存在，需要设置index_col=0或者指定其他列作为行索引
pd.read_csv('data_frame.csv', index_col=0)
```

```python
# 我们也可以使用read_table来读取, 需要指定数据分隔符
# 我们也可以通过将header=None,并且使用names指定自定义列索引名
# skiprows可以跳过前几行，或者指定哪些行skiprows=[1, 2, 3]
pd.read_table('data_frame.csv', 
              sep=',', 
              index_col=0, 
              header=None, 
              names=['d' + str(ix) for ix in range(1, 11)],
              skiprows=1
        )
```
