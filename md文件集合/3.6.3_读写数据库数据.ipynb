```python
from pandas import DataFrame,Series
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
from sqlalchemy.types import *


data_frame = DataFrame(
    np.random.randint(100, 1000, (10, 10)), 
    index=[ 'key' + str(ix) for ix in range(1, 11)], 
    columns=[ 'data' + str(ix) for ix in range(1, 11)],
)

data_frame.index.name = 'key'

data_frame
```

```python
# 创建数据库引擎
# mysql+pymysql 表示使用Mysql数据库的pymysql驱动
# root:chuanzhi@127.0.0.1/data_db 账号:密码,最后的data_db表示连接的数据库
engine = create_engine('mysql+pymysql://root:chuanzhi@127.0.0.1/data_db')

# 将DataFrame写入数据库
data_frame.to_sql(
    'my_data', 
    engine, 
    if_exists='replace',
    # 如果列不是默认的数字, 需要添加
    dtype={'key': VARCHAR(data_frame.index.get_level_values('key').str.len().max())}
)

# 从数据库中读取DataFrame
# set_index(['key']) 指定key这一列作为索引
# columns指定读取哪些字段信息
pd.read_sql('my_data', engine, columns=['key', 'data1', 'data3']).set_index(['key'])
# 按照条件从数据库中读取数据
pd.read_sql('select * from my_data where data5 > 300', engine).set_index(['key'])
```
