```python
from pandas import DataFrame, Series
import pandas as pd
import numpy as np

data_frame = DataFrame(
    np.random.randint(10, 100, (10, 3)), 
    columns=['data' + str(num) for num in range(1, 4)]
)

# 添加一列作为分组条件
data_frame['key'] = Series([list('ab')[np.random.randint(2)] for _ in range(10)], name='key')
data_frame
```

```python
# 对数据进行分组
data_by_group = data_frame.groupby('key')

# 打印分组数据
dict([x for x in data_by_group])['a']
```

```python
dict([x for x in data_by_group])['b']
```

```python
# 获取每一组最大的前三个元素
def top(data_frame):
    return data_frame.sort_values('data3')[-3:][::-1]

# 每一个元素加上10
def plus_ten(data_frame):
    return data_frame.iloc[:, :3] + 10

# 获取每一列的最大值
def get_max(data_frame):
    return data_frame.max()

# 将每一组分别应用apply函数
data_by_group.apply(top)
data_by_group.apply(plus_ten)
data_by_group.apply(get_max)
```
